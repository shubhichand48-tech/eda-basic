```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('House Price India.csv')
df.head()
```

## Basic Checks
**Explanation:** Run `df.info()`, `df.describe()` and check missing values.

**Interpretation:** Write observations after running the code.

```
df.info()
display(df.describe())
display(df.isnull().sum())
```

## Q1
**Explanation:** Find average, median and standard deviation of house prices.

**Interpretation:** Write observations after running the code.

```
print(df['Price'].mean())
print(df['Price'].median())
print(df['Price'].std())
```

## Q2
**Explanation:** Find the most common number of bedrooms.

**Interpretation:** Write observations after running the code.

```
print(df['number of bedrooms'].mode())
```

## Q3
**Explanation:** Check skewness of price.

**Interpretation:** Write observations after running the code.

```
print(df['Price'].skew())
plt.hist(df['Price']); plt.show()
```

## Q4
**Explanation:** Average price by bedrooms.

**Interpretation:** Write observations after running the code.

```
display(df.groupby('number of bedrooms')['Price'].mean())
```

## Q5
**Explanation:** Relationship between area (excluding basement) and price.

**Interpretation:** Write observations after running the code.

```
print(df['Price'].corr(df['Area of the house(excluding basement)']))
plt.scatter(df['Area of the house(excluding basement)'],df['Price'],s=5)
plt.show()
```

## Q6
**Explanation:** Find high-price low-area anomalies.

**Interpretation:** Write observations after running the code.

```
hp=df['Price'].quantile(.95)
la=df['Area of the house(excluding basement)'].quantile(.25)
display(df[(df['Price']>hp)&(df['Area of the house(excluding basement)']<la)])
```

## Q7
**Explanation:** Compare average price by floors and waterfront.

**Interpretation:** Write observations after running the code.

```
display(df.groupby(['number of floors','waterfront present'])['Price'].mean())
```

## Q8
**Explanation:** Minimum and maximum price.

**Interpretation:** Write observations after running the code.

```
print(df['Price'].min())
print(df['Price'].max())
```

## Q9
**Explanation:** Highest average price by Postal Code.

**Interpretation:** Write observations after running the code.

```
display(df.groupby('Postal Code')['Price'].mean().sort_values(ascending=False).head())
```

**Explanation:** Write at least five insights.

**Interpretation:** Write observations after running the code.

```
print('1. Three-bedroom houses dominate.\n2. Prices are positively skewed.\n3. Area positively affects price.\n4. Waterfront houses are costlier.\n5. Premium postal codes have highest average prices.')
```
